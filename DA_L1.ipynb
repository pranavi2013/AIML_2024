```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/root/.cache/kagglehub/datasets/awaiskaggler/insurance-csv/versions/1/insurance.csv")
df.head()
```

```python
#vector of ages[one dimensional continuous values]
age_vector = df['age'].tolist()
child_vector = df['children'].tolist()
print("Vector of Ages:", age_vector[:5])
```

```python
#factors[one dimensional categorical values]
smoker_factor = pd.Categorical(df['smoker'])
print(smoker_factor.categories)
```

```python
#List[one dimensional heterogeneous values(vectors+factors)]
insurance_data = [age_vector,smoker_factor]
print(insurance_data[:2])
```

```python
#matrices [two dimensional heterogeneous ]
matrix = np.array([df['age'].values, df['children'].values])
matrix = matrix.T
print(matrix[:5])
```

```python
#dataframes[two dimensional heterogeneous more than one row of lists]
new_df = pd.DataFrame({
    'Age': df['age'],
    'Smoker': df['smoker'],
    'Children': df['children']
})
new_df.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['smoker'], c='blue', alpha=0.6, label='Children')

plt.title("Scatter Plot of Age vs Smoker", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Smoker", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.scatter(df['age'], df['bmi'], c='blue', alpha=0.6, label='bmi')

plt.title("Scatter Plot of Age vs BMI", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Bmi", fontsize=12)

plt.tight_layout()
plt.show()
```

```python
filtered_df = df[df['expenses'] > 20000]
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['age'], filtered_df['expenses'], c='red', alpha=0.7, label='Filtered expenses')
plt.title("Scatter Plot of Age vs expense (Filtered: expenses > $20,000)", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Expenses", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# prompt: plot colour should be red

import matplotlib.pyplot as plt

plt.scatter(age_vector, smoker_factor,marker='+', color='red')
plt.xlabel("Age")
plt.ylabel("Smoker")
plt.title("Scatter Plot of Age(20-40) vs Smoker")
plt.show()
```

```python
a = df[df['age'] > 39]['smoker'].value_counts()
b = df[df['age'] <= 39]['smoker'].value_counts()

plt.figure(figsize=(8, 6))

plt.bar(['Above 39', 'Below 39'], [a.sum(), b.sum()], color=['skyblue', 'red'])

plt.xlabel("Age Group")
plt.ylabel("Number of Smokers")
plt.title("Number of Smokers Above and Below 39")

plt.show()
```
